```python
def calculo_media(nota1, nota2):
  media = (nota1 + nota2) // 2
  return media # Retorna o valor da função

nota1 = float(input("Digite uma nota: "))
nota2 = float(input("Digite outra nota: "))
# float = mostrar números com casas decimais

media = calculo_media(nota1, nota2)
print(f"Média: {media: .2f}")
# .2f = exibe duas casas decimais para float
```

## Exercício 33
Verificar número primo.

```python
numero = int(input("Digite um número: "))

contador = 0
divisor = 1
while(divisor <= numero):
  if numero % divisor == 0:
    contador += 1
  divisor += 1

if contador == 2:
  print("Número é primo")
else:
  print("Número não é primo")
```

## Exercício 34
Gerar lista de primos até N.

```python
def numero_primo(num):
    if num < 2:
        return False

    for i in range(2, num):
        if num % i == 0:
            return False

    return True


N = int(input("Digite um número: "))

primos = []

for numero in range(2, N + 1):
    if numero_primo(numero):
        primos.append(numero)

print(primos)
```

## Exercício 59
Crie uma função que remove duplicados de uma lista.

```python
def remove_duplicados(lista):
    nova_lista = []

    for n in lista:
        if n not in nova_lista:
            nova_lista.append(n)

    return nova_lista

numeros = [2, 2, 3, 8, 7, 7, 6, 5, 5, 4]

print(remove_duplicados(numeros))
```

## Exercício 80
Criar função que recebe lista de dicionários e retorna média de idades.

```python
def media_idades(pessoas):
    soma = 0

    for pessoa in pessoas:
        soma += pessoa["Idade"]

    return soma / len(pessoas)


pessoas = [
    {"Nome": "João", "Idade": 21},
    {"Nome": "Maria", "Idade": 20},
    {"Nome": "Ana", "Idade": 17}
]

print(f"A média de idades é: ", media_idades(pessoas))
```

## Exercício 81
Criar função que retorna maior idade.

```python
def maior_idade(pessoas):
    maior = pessoas[0]["Idade"]

    for pessoa in pessoas:
        if pessoa["Idade"] > maior:
            maior = pessoa["Idade"]

    return maior


pessoas = [
    {"Nome": "João", "Idade": 21},
    {"Nome": "Maria", "Idade": 20},
    {"Nome": "Ana", "Idade": 17}
]

print(f"A maior idade é: ", maior_idade(pessoas))
```

## Exercício 82

## Exercício 85
Inverter chave e valor de um dicionário.

```python
pessoa = {
    "João": 21,
    "Maria": 20,
    "Ana": 17
}

inversao = {}

for chave in pessoa:
    inversao[pessoa[chave]] = chave

print(inversao)
```

## Exercício 86
Contar frequência de letras em uma palavra usando dicionário.

```python
palavra = input("Digite uma palavra: ")

frequencia = {}

for letra in palavra:
    if letra in frequencia:
        frequencia[letra] += 1
    else:
        frequencia[letra] = 1

print(f"A frequência das letras na palavra é: ", frequencia)
```

## Exercício 87
Contar frequência de palavras em frase.

```python
frase = input("Digite uma frase: ")

frequencia = {}

palavras =  frase.split() # split separa a string em partes

for palavra in palavras:
    if palavra in frequencia:
        frequencia[palavra] += 1
    else:
        frequencia[palavra] = 1

print(f"A frequência das palavras na frase é: ", frequencia)
```

## Exercício 88
Agrupar números pares e ímpares em dicionário.

```python
numeros = [2, 7, 10, 13, 15, 20]

resultado = {
    "pares": [],
    "impares": []
}

for n in numeros:
    if n % 2 == 0:
        resultado["pares"].append(n)
    else:
        resultado["impares"].append(n)

print(resultado)
```

## Exercício 89
Criar função que verifica se todos valores são positivos.

```python
def valores_positivos(lista):
    for n in lista:
        if n <= 0:
            return False
    return True

numeros = [-2, -1, 2, 7, 10, 13, 15, 20]

print(f"A afirmação: 'Todos os valores são positivos' é: ", valores_positivos(numeros))
```

## Exercício 90
Criar função que retorna apenas valores únicos de lista.

```python
# Números que não se repetem
def valores_unicos(lista):
    nova_lista = []

    for n in lista:
        if n not in nova_lista:
            nova_lista.append(n)

    return nova_lista


numeros = [1, 1, 2, 3, 4, 4, 5, 5]

print(valores_unicos(numeros))
```

## Exercício 91
Criar função que recebe lista e retorna apenas pares.

```python
def numeros_pares(lista):
    pares = []

    for n in lista:
        if n % 2 == 0:
            pares.append(n)

    return pares


numeros = [1, 2, 3, 4, 5]

print(f"Lista com números pares: ", numeros_pares(numeros))
```

## Exercício 92
Criar função que soma lista usando reduce ou loop.

```python
def soma_lista(lista):
    soma = 0

    for n in lista:
        soma += n

    return soma


numeros = [1, 2, 3, 4, 5]

print(f"A soma da lista é: ", soma_lista(numeros))
```

## Exercício 93
Criar função que multiplica todos valores da lista.

```python
def multiplicacao_lista(lista):
    resultado = 1

    for n in lista:
        resultado *= n

    return resultado


numeros = [1, 2, 3, 4, 5]

print(f"O resultado da multiplicação da lista é: ", multiplicacao_lista(numeros))
```

## Exercício 94
Criar função que retorna lista invertida.

```python
def inversao_lista(lista):
    invertido = []

    for n in lista:
        invertido.insert(0, n)

    return invertido


numeros = [1, 2, 3, 4, 5]

print(inversao_lista(numeros))
```

## Exercício 95
Criar função que verifica se lista está ordenada.

```python
def esta_ordenada(lista):
    for i in range(len(lista) - 1):
        if lista[i] > lista[i + 1]:
            return False
    return True


numeros = [1, 2, 3, 4, 5]

print(f"A afirmação = 'A lista está ordenada' é:", esta_ordenada(numeros))
```

## Exercício 96
Criar função que remove valores negativos.

```python
def remove_negativos(lista):
    nova_lista = []

    for n in lista:
        if n >= 0:
            nova_lista.append(n)

    return nova_lista


numeros = [-3, -2, -1, 0, 1, 2, 3]

print(remove_negativos(numeros))
```

## Exercício 98
Criar função que retorna interseção de listas.

```python
def intersecao(lista1, lista2):
    resultado = []

    for n in lista1:
        if n in lista2:
            resultado.append(n)

    return resultado


lista1 = [2, 3, 5, 6]
lista2 = [5, 6, 7, 8]

print(intersecao(lista1, lista2))
```

## Exercício 99
Criar função que retorna diferença entre listas.

```python
def diferenca(lista1, lista2):
    resultado = []

    for n in lista1:
        if n not in lista2:
            resultado.append(n)

    return resultado


lista1 = [2, 3, 5, 6]
lista2 = [5, 6]

print(diferenca(lista1, lista2))
```

## Exercício 100
Criar função que conta elementos únicos.

```python
def contar_unicos(lista):
    return len(set(lista))


numeros = [1, 1, 2, 3, 3, 4, 5, 5, 6]

print(contar_unicos(numeros))
```
