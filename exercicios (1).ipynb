## Exercício 2
Receba duas notas e calcule a média.

```python
def calculo_media(nota1, nota2):
  media = (nota1 + nota2) // 2
  return media # Retorna o valor da função

nota1 = float(input("Digite uma nota: "))
nota2 = float(input("Digite outra nota: "))
# float = mostrar números com casas decimais

media = calculo_media(nota1, nota2)
print(f"Média: {media: .2f}")
# .2f = exibe duas casas decimais para float
```

## Exercício 3
Receba um número e mostre o dobro.

```python
def calculo_dobro(numero):
  dobro = (numero) * 2
  return dobro

numero = int(input("Digite um número: "))
dobro = calculo_dobro(numero)
print(f"O dobro desse número é: {dobro}")
```

## Exercício 4
Receba um número e mostre o triplo.

```python
def calculo_triplo(numero):
  triplo = (numero) * 3
  return triplo

numero = int(input("Digite um número: "))
triplo = calculo_triplo(numero)
print(f"O triplo, desse número é: {triplo}")
```

## Exercício 6
Receba um número e diga se é par ou ímpar.

```python
# Dúvida no exercício

def calculo_par_impar(numero):
  if numero // 2 == 0:
    return f"{numero} é par"
  else:
    return f"{numero} é ímpar"

numero = int(input("Digite um número: "))
print(calculo_par_impar(numero))
```

## Exercício 7
Receba três números e mostre o maior.

```python
numero1 = float(input("Digite o primeiro número: "))
numero2 = float(input("Digite o segundo número: "))
numero3 = float(input("Digite o terceiro número: "))
maior = numero1

if numero2 > maior:
    maior = numero2

if numero3 > maior:
    maior = numero3

print("O maior número é:", maior)
```
